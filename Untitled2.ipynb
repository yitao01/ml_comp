### Python 101: FizzBuzz

You are given an array of numbers. For each number in the array, you are asked to print `fizz` if the number is divisible by 5, `buzz` if the number is divisible by 3 and `fizzbuzz` if it is divisible by both 3 and 5. In all these cases, print the number as well.

```python
class solution (object):
  def fizzbuzz(self,n):
      ans=[]
      for num in range (1,n+1):
        d_3=(num%3==0)
        d_5=(num%5==0)
        res=""
        if d_3:
          res+="fizz"
        if d_5:
          res+="buzz"
        if not res:
          res=str(num)
        ans.append(res)
      return ans
solution().fizzbuzz(30) 
```

### Smoothing:

You are given an array containing some measurements from a sensor. You are asked to plot a smoothed version of the time series. Your function needs to take two inputs:
1. Time series as an array
2. Smoothing window (assume the window size is an odd number)

Assume the window is symmetric around the sample.

```python
import numpy as np
import matplotlib.pyplot as plt
def smoothing (nums, n):
  for i in range (len(nums)):
    sum=0
    for k in range (n):
      if (i+k<len(nums)):
        sum+=nums[i+k]
    nums[i]=sum/n
  return nums
x=np.linspace(1,501)
nums=np.cos(x)+np.random.normal(scale=0.5,size=len(x))
plt.plot(x,nums)
a=smoothing(nums,5)
plt.plot(x,a)
```
